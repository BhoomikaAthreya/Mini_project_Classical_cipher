```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter

# Load the original dataset
df = pd.read_csv('/content/sorted_caesar_cipher_dataset.csv')  # Make sure this is the path to your original dataset

# Feature 1: Character Frequency
char_vectorizer = CountVectorizer(analyzer='char')
char_freq_matrix = char_vectorizer.fit_transform(df['Cipher'])
char_freq_df = pd.DataFrame(char_freq_matrix.toarray(), columns=char_vectorizer.get_feature_names_out())

# Add character frequency features to the dataset
for col in char_freq_df.columns:
    df[f'char_freq_{col}'] = char_freq_df[col]

# Feature 2: Bigram Frequency
bigram_vectorizer = CountVectorizer(analyzer='char', ngram_range=(2, 2))
bigram_matrix = bigram_vectorizer.fit_transform(df['Cipher'])
bigram_df = pd.DataFrame(bigram_matrix.toarray(), columns=bigram_vectorizer.get_feature_names_out())

# Add bigram frequency features to the dataset
for col in bigram_df.columns:
    df[f'bigram_freq_{col}'] = bigram_df[col]

# Feature 3: Text Length
df['text_length'] = df['Cipher'].apply(len)

# Feature 4: Unique Character Count
df['unique_chars'] = df['Cipher'].apply(lambda x: len(set(x)))

# Save the updated dataset with new features
df.to_csv("encrypted_data_with_features.csv", index=False)

print("Updated dataset saved as 'encrypted_data_with_features.csv'")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import joblib

# Function to generate Caesar cipher dataset for all keys
def generate_caesar_dataset(plaintexts, max_shift=25):
    data = []
    for plaintext in plaintexts:
        for shift in range(max_shift + 1):  # Shift from 0 to 25
            ciphertext = ''.join(chr((ord(char) - ord('A') + shift) % 26 + ord('A')) for char in plaintext.upper() if char.isalpha())
            data.append((ciphertext, shift))
    return pd.DataFrame(data, columns=['ciphertext', 'key'])

# Example plaintexts
plaintexts = ["HELLO", "BAT", "CAESAR", "CIPHER", "SECURITY"]

# Generate the dataset
df = generate_caesar_dataset(plaintexts)

# Convert ciphertexts to features
def extract_features(ciphertext):
    features = np.zeros(26)
    for char in ciphertext:
        if char.isalpha():
            features[ord(char.upper()) - ord('A')] += 1
    return features

# Prepare the training data
X_train = np.array([extract_features(c) for c in df['ciphertext']])
y_train = df['key'].values

# Scale the training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Train the model
model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
model.fit(X_train_scaled, y_train)

# Save the trained model
joblib.dump(model, 'model.pkl')
joblib.dump(scaler, 'scaler.pkl')

# Function to decrypt the Caesar cipher text
def decrypt_caesar_cipher(ciphertext, key):
    decrypted_text = ""

    for char in ciphertext:
        if char.isalpha():  # Check if the character is a letter
            # Shift character back by the key
            shifted = ord(char) - key

            # Wrap around if the letter goes before 'A'
            if shifted < ord('A'):
                shifted += 26

            decrypted_text += chr(shifted)
        else:
            decrypted_text += char  # Keep spaces and non-alpha characters unchanged

    return decrypted_text

# Example ciphertext
ciphertext_example = "CBU"

# Extract features from the ciphertext
calculated_features = extract_features(ciphertext_example)

# Scale the features using the fitted scaler
calculated_features = scaler.transform(calculated_features.reshape(1, -1))

# Make the prediction
predicted_key = model.predict(calculated_features)
predicted_key_value = predicted_key[0]
print(f"Predicted Key for '{ciphertext_example}': {predicted_key_value}")

# Decrypt the ciphertext using the predicted key
decrypted_text = decrypt_caesar_cipher(ciphertext_example, predicted_key_value)
print(f"Decrypted Text: '{decrypted_text}'")
```
